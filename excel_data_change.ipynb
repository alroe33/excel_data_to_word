```python
from google.colab import files
import pandas as pd

# 엑셀 파일 업로드
uploaded = files.upload()

# 업로드한 엑셀 파일명 가져오기
filename = next(iter(uploaded))

# 엑셀 파일 로드
df = pd.read_excel(filename)

# O열 데이터 수정
df["O"] = df["O"].str.replace(":", "")

# 수정된 데이터 저장
df.to_excel("수정본.xlsx", index=False)

# 수정된 파일 다운로드
files.download("수정본.xlsx")
```
